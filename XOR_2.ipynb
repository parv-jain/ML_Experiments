```python
import numpy as np
import keras
from keras.layers import Dense, Input, Activation
from keras.models import Sequential
from keras.optimizers import SGD
```

```python
# One-hot encoding
```

```python
x = np.array([
        [0., 0.],
        [0., 1.],
        [1., 0.],
        [1., 1.]
       ])
y = np.array([[1., 0.], [0., 1.], [0., 1.], [1., 0.]])
y = np.array([[0.], [1.], [1.], [0.]])
```

```python
y = np.array([0., 0., 1., 1])
```

```python
y.reshape(-1, 1)
```

```python
x.shap
```

```python
model = Sequential()
model.add(Dense(4, input_dim=2))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('relu'))
```

```python
model.summary()
```

```python
model.compile(optimizer=SGD(lr=0.1, momentum=0.5), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x, y, verbose=1, epochs=1000, batch_size=None)
```

```python
model.get_weights()
```

```python
model.predict(x)
```
