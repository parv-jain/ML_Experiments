```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
xs = np.random.randn(100)
print (xs)
```

```python
a = -2.2
b = 3.1
```

```python
ys = a * xs + b
```

```python
plot.plot(xs, ys)
```

```python
w1 = 0.1
w2 = 0.1
```

```python
yhats = w1 * xs + w2
```

```python
plot.plot(xs, ys, xs, yhats)
```

```python
LR = 1e-3
```

```python
for i in range(10):
    dlw2 = 2 * (w1 * xs + w2 - ys).sum()
    dlw1 = 2 * ((w1 * xs + w2 - ys) * xs).sum()    
    w1 = w1 - LR * dlw1
    w2 = w2 - LR * dlw2
    print(w1,w2)
```

```python
plot.plot(xs, ys, xs, w1 * xs + w2)
```

```python
def loss(ys, yhats):
    return np.square(yhats - ys).sum()
losses1 = list()
w1s = list()
for i in np.linspace(-50, 50, 1000):
    yhats = i * xs + w2
    losses1.append(loss(ys, yhats))
    w1s.append(i)
    
losses2 = list()
w2s = list()
for i in np.linspace(-50, 50, 1000):
    yhats = w1 * xs + i
    losses2.append(loss(ys, yhats))
    w2s.append(i)
    
plot.plot(w1s, losses1)
plot.plot(w2s, losses2)
```
