```python
%matplotlib inline
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plot
```

```python
xs = np.random.randn(100)
```

```python
a = 8.5
b = -3.1
```

```python
ys = a * xs + b
```

```python
plot.plot(xs, ys)
```

```python
w1 = 0.1
w2 = 0.2
```

```python
yhats = w1 * xs + w2
```

```python
plot.plot(xs, ys, xs, yhats)
```

```python
def loss(ys, yhats):
    return np.square(yhats - ys).sum()
```

```python
loss(yhats, ys)
LR = 1e-5
w1 = 0.1
w2 = 0.2
# 8.5
# -3.1
```

```python
for i in range(1000):
    dlw2 = 2 * (w1 * xs + w2 - ys).sum()
    dlw1 = 2 * ((w1 * xs + w2 - ys) * xs).sum()
    w1 = w1 - LR * dlw1
    w2 = w2 - LR * dlw2
    #print (w1, w2)
```

```python
plot.plot(xs, ys, xs, w1 * xs + w2)
```

```python
w1s = np.linspace(-50, 50, 1000)
```

```python
losses = list()
w1s = list()
for i in np.linspace(-50, 50, 1000):
    yhat = i * xs + w2
    losses.append(loss(yhat, ys))
    w1s.append(i)
```

```python
plot.plot(w1s, losses)
```

```python
losses2 = list()
w2s = list()
for i in np.linspace(-50, 50, 1000):
    yhat = w1 * xs + i
    losses2.append(loss(yhat, ys))
    w2s.append(i)
```

```python
plot.plot(w1s, losses, w2s, losses2)
```
